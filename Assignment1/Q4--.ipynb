```python
from nltk.corpus import nps_chat as nps
```

```python
nps.fileids()
```

```python
posts = nps.xml_posts()    # Get all posts content from all nps files.
users = []
for post in posts:
    user = post.get('user')
    users.append(user)
users
```

```python
count = 0
for u in users:
    if 'user121' in u.lower():
        count += 1
count
```
